```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import seaborn as sns  
```

```python
# create a data matrix for the heatmap
data = np.random.rand(10,10)
ax = sns.heatmap(data, cmap='viridis',cbar=False)
ax.set_xticks([])
ax.set_yticks([])
```

```python
# create a data matrix for the heatmap
data = np.random.rand(10,10)
ax = sns.heatmap(data, cmap = 'rocket',cbar=False)
ax.set_xticks([])
ax.set_yticks([])
```

```python
data = np.random.choice([0,1],(10,10))
ax = sns.heatmap(data, cmap = 'Greys',cbar=False,linecolor='black',linewidth=0.1)
ax.set_xticks([])
ax.set_yticks([])
```
